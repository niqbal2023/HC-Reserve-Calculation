```python
import numpy as np
import matplotlib.pyplot as plt

def volumetric_oil_reserves(area, net_pay, porosity, oil_saturation, water_saturation, formation_vol_factor):
    """
    Calculate Original Oil in Place (OOIP) using the Volumetric Method.
    """
    OOIP = 7758 * area * net_pay * porosity * oil_saturation * (1 - water_saturation) / formation_vol_factor
    return OOIP

def exponential_decline(qi, Di, time):
    """
    Calculate production rate over time using exponential decline curve analysis.
    """
    qt = qi * np.exp(-Di * time)
    return qt

def cumulative_production(qi, Di):
    """
    Calculate ultimate recovery using exponential decline.
    """
    EUR = qi / Di
    return EUR
```

```python
# Example Inputs
area = 500  # acres
net_pay = 50  # feet
porosity = 0.2  # fraction
oil_saturation = 0.8  # fraction
water_saturation = 0.2  # fraction
formation_vol_factor = 1.2  # reservoir barrels per stock tank barrel

# Calculate OOIP
ooip = volumetric_oil_reserves(area, net_pay, porosity, oil_saturation, water_saturation, formation_vol_factor)
print(f"Original Oil in Place (OOIP): {ooip:.2f} STB")

# Plot OOIP
plt.figure(figsize=(6, 4))
plt.bar(['OOIP'], [ooip], color='green')
plt.ylabel('Stock Tank Barrels (STB)')
plt.title('Original Oil in Place (OOIP)')
plt.grid(axis='y')
plt.show()

# Decline Curve Analysis Example
qi = 1000  # Initial production rate (barrels per day)
Di = 0.1  # Decline rate per year
time = np.arange(0, 20, 1)  # Time in years

# Calculate production decline
declined_rates = exponential_decline(qi, Di, time)
EUR = cumulative_production(qi, Di)
print(f"Estimated Ultimate Recovery (EUR) using exponential decline: {EUR:.2f} STB")

# Plot decline curve
plt.figure(figsize=(8, 5))
plt.plot(time, declined_rates, label='Production Rate', color='b')
plt.xlabel('Time (Years)')
plt.ylabel('Production Rate (STB/day)')
plt.title('Exponential Decline Curve')
plt.legend()
plt.grid()
plt.show()
```
